``

``

````python
#insert your filepaths
filepath1 = "../images/cats/108_252738.jpg"
filepath2 = "../images/cats/smol_cat.jpg"
image1 = np.array(Image.open(filepath1))
image2 = np.array(Image.open(filepath2))
````

````python
#Extract the width and height of image1 and image2, this will be useful when placing it later
image1_width = image1.shape[0]
image1_height = image1.shape[1]

image2_width = image2.shape[0]
image2_height = image2.shape[1]
````

````python
#Calculate these points which determine where image2 will be inserted
x_axis_start = (image1_width - image2_width) // 2
x_axis_end = x_axis_start + image2_width

y_axis_start = (image1_height - image2_height) // 2
y_axis_end = y_axis_start + image2_height
````

````python
#use coordinates above to copy image2 onto the centre of image1
image1[x_axis_start:x_axis_end, y_axis_start:y_axis_end] = image2
Image.fromarray(image1)
````
